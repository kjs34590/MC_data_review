- 감성분석(VADER) - NLTK
- 리뷰 문장 단어 중에서 good +0.1, awful -0.1, perfect +0.2

```python
import nltk
from nltk import tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()
```

```python
try : 
    html = urlopen('http://www.imdb.com/title/tt4154756/reviews?ref_=tt_ql_3')
    # print(html)
except HTTPError as he :
    print('http error')
except URLError as ue :
    print('url error')
else : 
    soup = BeautifulSoup(html.read(), 'html.parser')
    print(soup)
```

```python
review_list = soup.find_all('div' , {'class' : 'imdb-user-review'})
review_list
```

```python
data = []
print('워드클라우드를 띄울 때 모든 리뷰 텍스트를 합친 문자열 sum_review : \n')
print('nltk sid - polarity_scores() : 단어의 긍정, 부정, 중립에 대한 점수 계산해서 종합점수(compound) 산출 \n')
sum_review = ''
for review in review_list :
    
    content = review.find('div', {'class' : 'text show-more__control'}).get_text()
    # print(content)
    # print("*" * 80)
    sum_review = sum_review+content
    
    line_list = tokenize.sent_tokenize(content) # 문장 단위 tokenize
    # print(line_list)
    # print("*" * 80)
    
    sum = 0.0
    total_sum = 0.0
    for sen in line_list :
        ss = sid.polarity_scores(sen)
        # print(ss)
        sum = sum + ss['compound']
    #print('sum : ', sum)
    total_sum = sum/len(line_list)
    print(total_sum)
```
