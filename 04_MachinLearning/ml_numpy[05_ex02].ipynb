```python
import numpy  as np
import pandas as pd
```

```python
# 1. 각 파일을 pd.read_csv() 로 읽어들인다

rating_row = pd.read_csv('./data/user_ratings.dat' , sep='::' , header=None)
movie_row = pd.read_csv('./data/movies.dat' , sep='::' , header=None)
user_row = pd.read_csv('./data/users.dat' , sep='::' , header=None)
```

```python
print('type - '   , type(rating_row) , type(movie_row) , type(user_row))
# print('head() - \n' , rating_row.head())
```

```python
# 2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가
rating_row.columns = ['userId' ,  'movieId' , 'rating', 'timestamp' ] 
movie_row.columns  = ['movieId' , 'title', 'genres' ] 
user_row.columns   = ['userId' ,  'gender' , 'age', 'drop1', 'drop2' ] 
```

```python
del user_row['drop1']
del user_row['drop2']
```

```python
# 3. 공통 컬럼을 이용한 데이터 병합
# merge()
rating_movie_merge = pd.merge(rating_row , movie_row , on='movieId' , how = 'left')
# rating_movie_merge
user_rating_movie_merge = pd.merge(rating_movie_merge , user_row , on='userId' , how = 'left')
user_rating_movie_merge
```

```python
# 4. user 평점 갯수 순위 확인(sort)
# user_rating_movie_merge['userId'].value_counts()
temp_dataframe = pd.DataFrame({
    'user_id_cnt' : user_rating_movie_merge['userId'].value_counts()
})
temp_dataframe.index.name = 'userId'
temp_dataframe = temp_dataframe.reset_index()

temp_dataframe['user_id_cnt_rank'] = temp_dataframe['user_id_cnt'].rank(ascending=False).astype(int)

result_merge = pd.merge(user_rating_movie_merge , temp_dataframe , on='userId' , how = 'left')
display( result_merge )
```

```python
# 7. 평점이 200개 이상인 영화들의 성별 평균 평점
temp_dataframe = pd.DataFrame({
    'movie_id_cnt' : result_merge['movieId'].value_counts()
})
temp_dataframe.index.name = 'movieId'
temp_dataframe = temp_dataframe.reset_index()
result_movie_merge = pd.merge(result_merge , temp_dataframe , on='movieId' , how = 'left')
result_movie_merge

over_200_movie = result_movie_merge[ result_movie_merge['movie_id_cnt'] >= 200 ]
# display( over_200_movie.head() )
gender_m = over_200_movie[ over_200_movie['gender'] == 'M']['rating'].mean()
gender_f = over_200_movie[ over_200_movie['gender'] == 'F']['rating'].mean()

print('남자 성별 평균 평점 - ',gender_m)
print('여자 성별 평균 평점 - ',gender_f)

```
