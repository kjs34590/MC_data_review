```python
# 양적 자료의 데이터 분포 확인(boxplot)
print('양적 자료의 데이터 분포 확인')
dataset[['displ', 'cyl', 'cty', 'hwy']].boxplot()
```

```python
# 질적 자료에 대한 데이터 빈도 확인
# 제조회사, 모델명, 변속기 종류, 구동방식, 연료종류, 자동차 종류
print('질적 자료의 데이터 빈도 확인')
dataset[['manufacturer']].value_counts()
dataset[['model']].value_counts()
dataset[['trans']].value_counts()
dataset[['drv']].value_counts()
dataset[['fl']].value_counts()
dataset[['class']].value_counts()
```

```python
# 문제01]
# 자동차 배기량(displ)에 따라 고속도로 연비(hwy)가 다른지를 알아보자
# 배기량이 4 이하인 자동차화 5이상인 자동차 중 어떤 자동차의 고속도로연비가 평균적으로 높은지 알아보자
# dataframe - query()

print('< hvy 4 under displ mean > ', '\n', dataset[dataset['displ'] <= 4].mean() ), print()
print('<hvy 5 over displ mean > ', '\n', dataset[dataset['displ'] >= 5].mean() )
print()

print('< hvy 4 under displ mean > ', '\n', dataset.query('displ <= 4').mean() ), print()
print('<hvy 5 over displ mean > ', '\n', dataset.query('displ >= 5').mean() )
```

```python
# 문제 02]
# 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다. 
# audi와 toyota 중 어느 manufacturer(자동차 제조 회사)의 cty(도시연비)가 평균적으로 더 높은지 알아보시오.

# display(dataset.groupby('manufacturer').agg(np.mean).loc[:,'cty'])
# dataset.pivot_table(index= 'manufacturer', values= 'cty', aggfunc= 'mean', margins= True)

print('audi - ', dataset[dataset['manufacturer']=='audi']['cty'].mean)
print('audi - ', dataset[dataset['manufacturer']=='toyota']['cty'].mean)
```

```python
# 문제 03]
# chevrolet, ford, honda 자동차의 고속도로 연비 평균을 알아보려고 한다. 
# 이 회사들의 데이터를 추출한 후 hwy 전체 평균을 확인하시오.

print('chevrolet, ford, honda - ', dataset[(dataset['manufacturer'] == 'chevrolet') |
                                           (dataset['manufacturer'] == 'ford') |
                                           (dataset['manufacturer'] == 'honda')]['hwy'].mean())
print()

# query를 쓰는 방법
manufacturer_list = ['chevrolet', 'ford', 'honda']
print('query - ', dataset.query('manufacturer in @manufacturer_list')['hwy'].mean())
```

```python
# 연료가격 변수 추가 (price_fl)
print('연료가격 변수 추가')
print('c : 가스 2.35, d : 디젤 2.38, e ; 에탄올 2.11, p : 고급휘발유 2.76, r : 보통휘발유 : 2.22')

temp_frm = pd.DataFrame({
    'fl' : ['c', 'd', 'e', 'p', 'r'],
    'price_fl' : [2.35, 2.38, 2.11, 2.76, 2.22]
})
temp_frm
```

```python
merge_dataset = pd.merge(dataset, temp_frm, on='fl', how='inner')
merge_dataset
```
