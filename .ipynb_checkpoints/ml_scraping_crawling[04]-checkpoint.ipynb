- 1~25 page 내용 추출하기 [실습]

```python
driver = webdriver.Chrome(path)
driver.get('https://datalab.naver.com/shoppingInsight/sCategory.naver')

next_tag = driver.find_element_by_css_selector('a.btn_page_next')

shop_list = []

for i in range(1,26) : 
    ul_tag = driver.find_element_by_css_selector('ul.rank_top1000_list')
    li_tags = ul_tag.find_elements_by_tag_name('li')
    for li in li_tags :
        shop_list.append(li.text.split('\n'))

    next_tag.click()
```
