#### DecisionTree에서 사용할 수 있는 파라미터

- criterion : 분할 품질을 측정하는 기능 (default : gini)
- splitter : 각 노드에서 분할을 선택하는 데 사용되는 전략 (default : best)
- max_depth : 트리의 최대 깊이 (값이 클수록 모델의 복잡도가 올라간다.)
- min_samples_split : 자식 노드를 분할하는데 필요한 최소 샘플 수 (default : 2)
- min_samples_leaf : 리프 노드에 있어야 할 최소 샘플 수 (default : 1)
- min_weight_fraction_leaf : min_sample_leaf와 같지만 가중치가 부여된 샘플 수에서의 비율
- max_features : 각 노드에서 분할에 사용할 특징의 최대 수
- random_state : 난수 seed 설정
- max_leaf_nodes : 리프 노드의 최대수
- min_impurity_decrease : 최소 불순도
- min_impurity_split : 나무 성장을 멈추기 위한 임계치
- class_weight : 클래스 가중치
- presort : 데이터 정렬 필요 여부

```python
# 데이터 로드
iris = load_iris()

# 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(iris.data  , 
                                                    iris.target ,
                                                    test_size = 0.2 , 
                                                    random_state = 200 ) 
# 의사결정트리 
dtc = DecisionTreeClassifier(random_state = 100) 
dtc.fit(X_train , y_train)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(dtc,
                out_file = 'iris_tree.dot',
                class_names = iris.target_names,
                feature_names = iris.feature_names,
                impurity = True,
                filled = True)
```

```python
import os

os.environ['PATH'] += os.pathsep+'C:/Program Files/Graphviz/bin/'

with open('iris_tree.dot') as file :
    dot_graph = file.read()
    
graphviz.Source(dot_graph)
```

```python
gscv_dtc = DecisionTreeClassifier()

param = {
    'criterion' : ['gini', 'entropy'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3],
    'min_samples_split' : [2,3]
}

grid_gscv_tree = GridSearchCV(gscv_dtc,
                             param_grid = param,
                             cv = 3,
                             refit = True) # 최적의 하이퍼 파라미터를 찾아 재학습 
```

```python
grid_gscv_tree.fit(X_train, y_train)
grid_gscv_tree.cv_results_
```

```python
score_frm = pd.DataFrame(grid_gscv_tree.cv_results_)
score_frm
```

```python
score_frm[['params', 'rank_test_score']]
```

```python
print('최적 파라미터')
print(grid_gscv_tree.best_params_)
print()
print('높은 정확도')
print(grid_gscv_tree.best_score_)
```

```python
estimator = grid_gscv_tree.best_estimator_
y_predict = estimator.predict(X_test)
print('answer : \n', y_test)
print()
print('guess : \n', y_predict)
print()
print('테스트 정확도 : ', accuracy_score(y_test, y_predict))
```

```python
# 학습 / 테스트 데이터세트 분리
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                   test_size=0.3)

# DecisionTreeClassifier 객체 생성
dtc = DecisionTreeClassifier()

# 학습 수행
dtc.fit(X_train, y_train)

# 학습된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행
pred = dtc.predict(X_test)

# 예측 성능 평가
print('예측 정확도 : {0:.4f}'.format(accuracy_score(y_test, pred)))
```

```python
# graphviz 시각화
export_graphviz(dtc,
                out_file = 'cancer.dot',
                class_names = cancer.target_names,
                feature_names = cancer.feature_names,
                impurity = True,
                filled = True)

with open('cancer.dot') as file :
    dot_graph = file.read()
    
graphviz.Source(dot_graph)
```

```python
# 교차 검증 1. Stratified KFold
```

```python
# 교차 검증 2. cross_val_score
scores = cross_val_score(dtc, cancer.data, cancer.target, scoring='accuracy', cv=5)
print('폴드 정확도 : ', scores, '\n평균 정확도 : ',np.round(np.mean(scores), 4))
```

```python
# 하이퍼 파라미터 튜닝
gscv_dtc = DecisionTreeClassifier()

param = {
    'criterion' : ['gini', 'entropy'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'min_samples_split' : [2,3,4]
}

grid_gscv_tree = GridSearchCV(gscv_dtc,
                             param_grid = param,
                             cv = 3,
                             refit = True) # 최적의 하이퍼 파라미터를 찾아 재학습 

grid_gscv_tree.fit(X_train, y_train)
grid_gscv_tree.cv_results_

print('최적 파라미터')
print(grid_gscv_tree.best_params_)
print()
print('높은 정확도')
print(grid_gscv_tree.best_score_)
```
