- 시도별 전출입 인구수를 이용한 라인 플롯 실습
- fillna(대체값 |'missing'|)

```python
dataset = pd.read_excel('C:/DATA_WORKS/PYTHON_WORKS/data/시도별 전출입 인구수.xlsx')
dataset.head()
```

```python
dataset = dataset.fillna(method='ffill')
dataset.head()
```

```python
print('서울에서 다른 지역으로 이동한 데이터만 추출하여 서브셋 만들기')

mask = (dataset['전출지별'] == '서울특별시') & (dataset['전입지별']!='서울특별시')
seoul_frm = dataset[mask]
seoul_frm.head()
```

```python
print('전출지별 삭제')
seoul_frm = seoul_frm.drop(['전출지별'], axis=1)
seoul_frm.head()
```

```python
print('전입지별 인덱스 이름을 전입지로 변경')
print('rename - ')
print()

seoul_frm.rename({'전입지별':'전입지'}, axis = 1, inplace = True)
seoul_frm.head()
```

```python
print('전입지를 인덱스로 변경')
print('set_index - ')
print()

# seoul_frm.set_index('전입지', inplace = True)
seoul_frm.head()
```

```python
# 1980년대부터 인덱싱하여 변수 저장
to_ggd_values = seoul_frm.loc['경기도'].values[11:]
to_ggd_index = seoul_frm.loc['경기도'].index[11:]
to_ic_values = seoul_frm.loc['인천광역시'].values[11:]
to_ic_index = seoul_frm.loc['인천광역시'].index[11:]

print(to_ggd_values,'\n', to_ggd_index,'\n', to_ic_values,'\n', to_ic_index,'\n')
type(to_ggd_values)
```

```python
# 서울 -> 타 지역 전입 데이터 기반 시각화
print('01. 경기도로 이동한 인구 데이터만 선택해 시각화')
plt.figure(figsize = (15, 5)) # 틀 만들기
plt.title('서울 -> 경인지역 전입 인구')

plt.plot(to_ggd_index, to_ggd_values,
         c = 'blue', 
        lw = 2, ls = '-',
        marker = 'o', markersize = 4, mec = 'blue')

plt.plot(to_ggd_index, to_ic_values,
         c = 'green', 
        lw = 2, ls = '-',
        marker = 'o', markersize = 4, mec = 'green')

plt.xlabel('연도')
plt.ylabel('이동인구 수')
plt.xticks(rotation = 'vertical')
plt.legend(labels = ['서울 -> 경기', '서울 -> 인천'], loc = 'best')

plt.show()
plt.close()
```

```python
print('subplot 이용한 시각화')
fig = plt.figure(figsize = (12, 5))
ax01 = fig.add_subplot(2, 1, 1) # (1, 2, 1)
ax02 = fig.add_subplot(2, 1, 2) # (1, 2, 2) 로 그리면 횡으로 그래프 배치

plt.title('서울 -> 경인지역 전입 인구')


ax01.plot(to_ggd_index, to_ggd_values, marker='+', markersize = 4)

ax01.set_xlabel('연도')
ax01.set_ylabel('이동인구 수')
ax01.set_xticklabels(labels = to_ggd_index, rotation = 'vertical')
ax01.legend(labels = ['서울 -> 경기'], loc = 'best')
ax01.set_ylim(5000, 700000)

ax02.plot(to_ic_index, to_ic_values, marker='+', markersize = 4)

ax02.set_xlabel('연도')
ax02.set_ylabel('이동인구 수')
ax02.set_xticklabels(labels = to_ic_index, rotation = 'vertical')
ax02.legend(labels = ['서울 -> 인천'], loc = 'best')
ax02.set_ylim(5000, 200000)

plt.show()
plt.close()
```

```python
print('충남, 경북, 강원으로 이동한 인구 데이터만 선택해서 하나의 figure에 출력하기')
print()

# 01. 데이터 변수 저장
index = seoul_frm.columns.values    # index = list(map(str, range(1970, 2018)))
to_cn_values = seoul_frm.loc['충청남도'].values
to_gb_values = seoul_frm.loc['경상북도'].values
to_kw_values = seoul_frm.loc['강원도'].values

print(to_cn_values,'\n', to_gb_values,'\n', to_kw_values,'\n')

# 02. 서울 -> 타 지역 전입 데이터 기반 시각화
print('서울 -> 충남/경북/강원으로 이동한 인구 데이터 시각화')
plt.figure(figsize = (15, 5)) # 틀 만들기
plt.title('서울 -> 충남/경북/강원 전입 인구')

plt.plot(index, to_cn_values,
         c = 'green', 
        lw = 2, ls = '-',
        marker = 'o', markersize = 6, mec = 'g')

plt.plot(index, to_gb_values,
         c = 'blue', 
        lw = 2, ls = '--',
        marker = '+', markersize = 6, mec = 'b')

plt.plot(index, to_kw_values,
         c = 'orange', 
        lw = 2, ls = '-.',
        marker = '|', markersize = 6, mec = 'orange')

plt.xlabel('연도')
plt.ylabel('이동인구 수')
plt.xticks(rotation = 'vertical')
plt.legend(labels = ['서울 -> 충남', '서울 -> 경북', '서울 -> 강원'], loc = 'best')

plt.show()
plt.close()
```

```python
print('조건 1. 충청남도, 경상북도, 강원도, 전라남도')
print('조건 2. 2010 - 2017년 데이터')

# 01. 데이터 변수 저장
index_1017 = list(map(str, range(2010, 2017)))
cn_1017_val = seoul_frm.loc['충청남도'].values[40:47]
gb_1017_val = seoul_frm.loc['경상북도'].values[40:47]
kw_1017_val = seoul_frm.loc['강원도'].values[40:47]
jn_1017_val = seoul_frm.loc['전라남도'].values[40:47]


print(to_cn_values,'\n', to_gb_values,'\n', to_kw_values,'\n')

# 02. 서울 -> 타 지역 전입 데이터 기반 시각화
print('2010~2017년 서울 -> 충남/경북/강원/전남으로 이동한 인구 데이터 시각화')
plt.figure(figsize = (15, 5)) # 틀 만들기
plt.title('2010~2017년 서울 -> 충남/경북/강원/전남 전입 인구')

plt.plot(index_1017, cn_1017_val,
         c = 'green', 
        lw = 2, ls = '-',
        marker = 'o', markersize = 6, mec = 'g')

plt.plot(index_1017, gb_1017_val,
         c = 'blue', 
        lw = 2, ls = '--',
        marker = '+', markersize = 6, mec = 'b')

plt.plot(index_1017, kw_1017_val,
         c = 'orange', 
        lw = 2, ls = '-.',
        marker = '|', markersize = 6, mec = 'orange')

plt.plot(index_1017, jn_1017_val,
         c = 'red', 
        lw = 2, ls = '-',
        marker = '|', markersize = 6, mec = 'red')

plt.xlabel('연도')
plt.ylabel('이동인구 수')
plt.xticks()
plt.legend(labels = ['서울 -> 충남', '서울 -> 경북', '서울 -> 강원', '서울 -> 전남'], loc = 'best')

plt.show()
plt.close()
```

## Bar plot 그리기

- bar plo(plt.bar(X, Y))
- X => 범주형 데이터

```python
# titanic, iris 로드
import seaborn as sns
iris_datasets = sns.load_dataset('iris')
titanic_datasets = sns.load_dataset('titanic')

survived_index = titanic_datasets.groupby('pclass').survived.sum().index
survived_values = titanic_datasets.groupby('pclass').survived.sum().values
print(survived_index, survived_values)
```

```python
print('titanic : 선실 등급에 따른 생존자 수 시각화')
print()

# 01. 데이터 변수 저장
survived_index = titanic_datasets.groupby('pclass').survived.sum().index
survived_values = titanic_datasets.groupby('pclass').survived.sum().values

print(to_cn_values,'\n', to_gb_values,'\n', to_kw_values,'\n')

# 02. Box plot 시각화
print('선실 등급에 따른 생존자 수')
plt.figure(figsize = (15, 5)) # 틀 만들기
plt.title('선실 등급에 따른 생존자 수')

plt.bar(survived_index, survived_values)

plt.xlabel('선실 등급')
plt.ylabel('생존자 수')
plt.legend(labels = ['생존자 수'], loc = 'best')

plt.show()
plt.close()
```

```python
# 02-1 horizontal bar
print('선실 등급에 따른 생존자 수')
plt.figure(figsize = (15, 5)) # 틀 만들기
plt.title('선실 등급에 따른 생존자 수')

plt.barh(survived_index, survived_values) # plt.barh() -> horizontal bar plot 생성
 
plt.xlabel('생존자 수') # x-y축 바꿔줘야 함
plt.ylabel('선실 등급')
plt.legend(labels = ['생존자 수'], loc = 'best')

plt.show()
plt.close()
```

```python
print('subset 만들기 : ')
print('조건 1. 충청남도, 경상북도, 강원도, 전라남도')
print('조건 2. 2010 - 2017년 데이터')

# 01. 조건에 맞는 subset 실행
subset_frm = seoul_frm.loc[['충청남도', '경상북도', '강원도', '전라남도'], '2010':'2017']
t_subset_frm = subset_frm.T
t_subset_frm
```

```python
print('연도에 따른 권역별 인구 전입 수 : ')


plt.figure()

t_subset_frm.plot(kind='bar', figsize = (15,5))

plt.title('연도에 따른 권역별 인구 전입 수')
plt.xlabel('연도')
plt.ylabel('전입인구 수')

plt.legend(labels = '연도별 전입인구 수', loc = 'best')

plt.show()
plt.close()
```

```python
print('전입인구 수의 합계를 구해서 새로운 열(합계)을 추가하고 - ')
print('합계에 대해 큰 값부터 정렬')
print('수평막대그래프로 시각화하기')

subset_frm['합계'] = subset_frm.sum(axis=1)
print(subset_frm['합계'], type(subset_frm['합계'])) # Series return된다.

subset_sorted = subset_frm[['합계']].sort_values(by='합계', ascending=True) #[] 를 추가해 DataFrame으로 변환

plt.figure()

subset_sorted.plot(kind='barh',
                  figsize = (20, 10),
                  color = ('blue', 'orange', 'green', 'red'))
plt.show()
plt.close()
```

```python
print('종을 기준으로 Grouping해 각 그룹 평균을 시각화')
print('조건 : sepal_length = 꽃받침길이, sepal_width = 꽃받침너비, petal_length = 꽃잎길이, petal_width = 꽃잎너비')
print('kind - bar')

iris_mean = iris_datasets.groupby('species').agg(np.mean)
iris_mean = iris_mean.rename(columns = {
    iris_mean.columns[0] : '꽃받침길이' ,
    iris_mean.columns[1] : '꽃받침너비' ,
    iris_mean.columns[2] : '꽃잎길이' ,
    iris_mean.columns[3] : '꽃잎너비'})
display(iris_mean, type(iris_mean))

iris_mean.plot(kind='barh', title = '붓꽃 종별 꽃잎, 꽃받침 사이즈 평균', fontsize = '20',
                figsize = (20, 10), 
                color = ('palegreen', 'seagreen', 'peachpuff', 'salmon'))
```

```python
# 방법 2 
print('종을 기준으로 Grouping해 각 그룹 평균을 시각화')
print('조건 : sepal_length = 꽃받침길이, sepal_width = 꽃받침너비, petal_length = 꽃잎길이, petal_width = 꽃잎너비')
print('kind - bar')

iris_frm = iris_datasets.rename(columns = {
    iris_datasets.columns[0] : '꽃받침길이' ,
    iris_datasets.columns[1] : '꽃받침너비' ,
    iris_datasets.columns[2] : '꽃잎길이' ,
    iris_datasets.columns[3] : '꽃잎너비' ,
    iris_datasets.columns[4] : 'Y'
})
iris_frm
```

```python
mean_iris_frm = iris_frm.groupby(iris_frm.Y).mean()
print(mean_iris_frm)
```

```python
plt.figure()

mean_iris_frm.T.plot(kind='bar',
                 figsize = (20, 10),
                 rot = 0)
plt.show()
plt.close()
```
