### barplot() : 범주형데이터에 따른 시각화(평균, 편차, 에러)

```python
titanic_dataset.head()
```

```python
fig = plt.figure(figsize=(20,6))

sns.set_style('dark')

area01 = fig.add_subplot(1, 3, 1)
area01.set_title('sex - survived')
sns.barplot(ax = area01, data=titanic_dataset, x='sex', y='survived')

area02 = fig.add_subplot(1, 3, 2)
area02.set_title('sex - survived option hue')
sns.barplot(ax = area02, data=titanic_dataset, x='sex', y='survived', hue='class')

area03 = fig.add_subplot(1, 3, 3)
area03.set_title('sex - survived option hue, dodge')
sns.barplot(ax = area03, data=titanic_dataset, x='sex', y='survived', hue='class', dodge = False)

plt.show()
plt.close()
```

```python
# 요일별 전체 팁과 요일별 수입 시각화
fig = plt.figure(figsize=(20,6))

area01 = fig.add_subplot(1,2,1)
area01.set_title('요일별 전체 팁')
sns.barplot(ax = area01, data=tips_dataset, x= 'day', y='tip')

area02 = fig.add_subplot(1,2,2)
area02.set_title('요일별 수입')
sns.barplot(ax = area02, data = tips_dataset, x= 'day', y='total_bill')

plt.show()
plt.close()
```

### countplot() : 빈도를 표시하는 함수

```python
fig = plt.figure(figsize=(20,6))

sns.set_style('dark')

area01 = fig.add_subplot(1, 3, 1)
area01.set_title('class')
sns.countplot(ax = area01, data=titanic_dataset, x='class')

area02 = fig.add_subplot(1, 3, 2)
area02.set_title('class - who')
sns.countplot(ax = area02, data=titanic_dataset, x='class',hue='who', palette='Set3')

area03 = fig.add_subplot(1, 3, 3)
area03.set_title('class - who, option - hue/dodge')
sns.countplot(ax = area03, data=titanic_dataset, x='class', hue='who', dodge = False, palette='Set3')

plt.show()
plt.close()
```

```python
# 요일별 전체 팁과 요일별 수입 시각화

plt.title('요일별 팁을 준 빈도')
sns.countplot(data=tips_dataset, x= 'day')

plt.show()
plt.close()
```

### stripplot() : 모든 데이터를 점으로 시각화
- jitter = True : overlapped 데이터는 겹치지 않게 표시하는 옵션

```python
# 요일별 전체 팁과 요일별 수입 시각화

plt.title('요일별 팁 횟수')
sns.stripplot(data=tips_dataset, x= 'day', y='total_bill', palette='Set3', jitter=True)

plt.show()
plt.close()
```

### swarmplot()

```python
plt.title('요일별 팁 횟수')
sns.swarmplot(data=tips_dataset, x= 'day', y='total_bill', palette='Set3') # jitter 제거

plt.show()
plt.close()
```

```python
# titanic
fig = plt.figure(figsize=(20,6))

sns.set_style('dark')

area01 = fig.add_subplot(1, 2, 1)
area01.set_title('age by class : strip')
sns.stripplot(ax = area01, data=titanic_dataset, x='class', y='age', jitter=True)

area02 = fig.add_subplot(1, 2, 2)
area02.set_title('age by class : swarm')
sns.swarmplot(ax = area02, data=titanic_dataset, x='class',y='age')

plt.show()
plt.close()
```

```python
# tips
# 요일과 성별에 따른 전체 팁 분포 확인

fig = plt.figure(figsize=(20,6))

area01 = fig.add_subplot(1,2,1)
area01.set_title('요일별/성별 팁 분포 - strip')
sns.stripplot(ax = area01, data=tips_dataset, x='day', y='tip', hue='sex')

area02 = fig.add_subplot(1,2,2)
area02.set_title('요일별/성별 팁 분포 - swarm')
sns.swarmplot(ax = area02, data = tips_dataset, x= 'day', y='tip', hue='sex')

plt.show()
plt.close()
```

### elem_school_folium.xlsx 실습

```python
school_frm = pd.read_excel('C:/DATA_WORKS/PYTHON_WORKS/multicampus/data/elem_school_folium.xlsx')
school_frm.head()
```

```python
# Feature Selection
# 학급명, 학교급구분, 소재지도로주소명, 위도, 경도
subset_frm = school_frm.filter(['학교명', '학교급구분', '소재지도로명주소', '위도', '경도'])
subset_frm
```

```python
print('조건 - 주소가 서울시, 학교급구분이 초등학교인 곳만 출력')
print('hint - str.contains')
# 결측값 처리
print('결측값 확인 : \n', subset_frm.isnull().sum())

# 결측값 drop
subset_frm.dropna(inplace=True)

# 재확인
print('재확인 : \n', subset_frm.isnull().sum())
```

```python
condition = "학교급구분=='초등학교' and 소재지도로명주소.str.contains('서울특별시').values"
filter_frm = subset_frm.query(condition)
display(filter_frm)
```

```python
seoul_school_map = g.Map(location=[37.5532035880752, 126.98739716743527], zoom_start=12,
           tiles = 'Stamen Terrain')

for idx in filter_frm.index : 
    g.Marker([filter_frm.loc[idx, '위도'], filter_frm.loc[idx, '경도']],
            popup = filter_frm.loc[idx, '학교명']).add_to(seoul_school_map)

seoul_school_map
```

```python
pop_frm = pd.read_excel('C:/DATA_WORKS/PYTHON_WORKS/multicampus/data/경기도인구데이터.xlsx')
pop_frm.set_index('구분', inplace=True)
pop_frm.head()
```

```python
import json
geo_data = json.load(open('C:/DATA_WORKS/PYTHON_WORKS/multicampus/data/경기도행정구역경계.json', encoding='utf-8'))
geo_data
```

- json 확인 : 

'feature' > 'properties' > 'name'

'features' > 'geometry' > 'coordinates'

```python
pop_frm.columns = pop_frm.columns.map(str)
pop_frm.columns
```

```python
gyeonggi_map = g.Map(location=[37.26385623844609, 127.02878746071707],
                    zoom_start=12)

print('구분된 경계선을 표시하는 객체 - ')

g.Choropleth(geo_data = geo_data,
             data = pop_frm['2017'],
             columns = [pop_frm.index, pop_frm['2017']],
             key_on = 'feature.properties.name').add_to(gyeonggi_map)

gyeonggi_map
```
