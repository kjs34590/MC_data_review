#### pivot

- 새로운 데이터프레임이 만들어진다
- pivot(index, columns, data values)
- pivot_table(data, values, index, columns, aggfunc = )

```python
import seaborn as sns
iris_datasets = sns.load_dataset('iris')
titanic_datasets = sns.load_dataset('titanic')
```

```python
titanic_datasets.head()
```

```python
pclass_sex_frm = titanic_datasets.loc[:, ['pclass', 'sex']]
pclass_sex_frm
```

```python
print('성별과 객실 등급에 따른 승객 수 집계')
print()
print('groupby : ')
pclass_sex_pivot_frm = pclass_sex_frm.groupby(['sex', 'pclass']).size().reset_index(name='cnt')
display(pclass_sex_pivot_frm)
print()
print('pivot : ')
pclass_sex_pivot_frm.pivot('sex', 'pclass', 'cnt')
print('pivot tavle : ')
display(titanic_datasets.pivot_table(index='sex', columns='pclass', values='cnt', aggfunc=np.sum))
```

```python
print('성별과 생존여부에 따른 집계수 확인')
sex_survived_frm = titanic_datasets.loc[:, ['sex', 'survived']]
display(sex_survived_frm)
print()

print('pivot : ')
sex_survived_pivot_frm = sex_survived_frm.groupby(['sex', 'survived']).size().reset_index(name='cnt')
display(sex_survived_pivot_frm)
print()

print('pivot table : ')
titanic_datasets['cnt'] = 1
display(titanic_datasets.pivot_table(index='sex', columns='survived', values='cnt', aggfunc=np.sum, margins = True, margins_name='분석결과'))
```

```python
pivot_frm.pivot_table(index='city', columns='year', values='pop', margins = True, margins_name = '집계')
```

- qcut() : 범주를 만든다 (동일한 개수)
- cut() : 범주를 만든다 (동일한 길이) 
-> 그룹에 대한 통계량 확인

```python
print('각 붓꽃 꽃잎의 길이가 해당 종 내에서 대-중-소 어느 그룹에 해당되는지 분류')
# display(iris_datasets)
iris_grp = iris_datasets.groupby('species').petal_length
for key, group in iris_grp :
    print('key - ', key)
    print('len - ', len(group))
    print()
    print(group)
    print('*' * 50)
```

```python
iris_datasets.groupby('species').petal_length.size()
```

```python
def cut_petal_length(s) :
    return pd.cut(s, 3, labels=['대', '중', '소'])
```

- transform() :데이터프레임 자체를 변환시켜주는 함수

```python
# iris_datasets['category'] = iris_datasets.groupby('species').petal_length.transform(cut_petal_length)
iris_datasets['category'] = iris_datasets.groupby('species').petal_length.apply(cut_petal_length)
# iris_datasets['category'] = iris_datasets.groupby('species').petal_length.agg(cut_petal_length)
```

```python
print('[문제01] - ')
print('qcut 명령으로 세 개의 나이 그룹을 만든다. - ')
titanic_datasets = sns.load_dataset('titanic')
titanic_datasets['age_category'] = pd.qcut(titanic_datasets['age'] , 3 , labels = ['A' , 'B' , 'C' ])
titanic_datasets
```

```python
print('[문제02] - ')
print('성별, 선실, 나이그룹(age_category) 의한 생존율을 데이터프레임으로 생성하고')
print('행에는 성별 및 나이그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다 -')
print('생존률은 해당 그룹의 생존 인원수를 전체 인원수로 나눈 값이다. -')
print('cut - pivot_table')

frm = titanic_datasets.pivot_table(index=['sex', 'age_category'],
                            columns = 'pclass', values = 'survived',
                            margins = True, margins_name = 'survived_ratio',
                            aggfunc = 'count')
display(frm)
print()
print('전체 생존자 수')
frm.loc['survived_ratio', 'survived_ratio']
print('그룹별 생존자 비율')
display(frm / frm.loc['survived_ratio', 'survived_ratio'][0])
# 백분율 변환
survived_percentage = round(frm / frm.loc['survived_ratio', 'survived_ratio'][0]*100)
display(survived_percentage)
```

```python
print('문제[03] - ')
print('성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다. - ')
print()
print('데이터 추출')
frm = titanic_datasets.pivot_table(index='sex',
                            columns = 'pclass', values = 'survived',
                            margins = True, margins_name = 'survived_ratio',
                            aggfunc = 'count')
display(frm)
print('answer')
survived_frm = frm / titanic_datasets.shape[0]
display(survived_frm)
```
