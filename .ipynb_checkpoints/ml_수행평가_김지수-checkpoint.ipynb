```python
import numpy  as np
import pandas as pd
```

### 1. 각 파일을 pd.read_csv() 로 읽어들인다

```python
# 1. 각 파일을 pd.read_csv() 로 읽어들인다

rating_row = pd.read_csv('./data/user_ratings.dat' , sep='::' , header=None)
movie_row = pd.read_csv('./data/movies.dat' , sep='::' , header=None)
user_row = pd.read_csv('./data/users.dat' , sep='::' , header=None)
```

```python
print('type - '   , type(rating_row) , type(movie_row) , type(user_row))
# print('head() - \n' , rating_row.head())
```

### 2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가

```python
# 2. 데이터 확인 및 구조분석과 각 데이터 프레임에 헤더를 추가
rating_row.columns = ['userId' ,  'movieId' , 'rating', 'timestamp' ] 
movie_row.columns  = ['movieId' , 'title', 'genres' ] 
user_row.columns   = ['userId' ,  'gender' , 'age', 'drop1', 'drop2' ] 
```

```python
del user_row['drop1']
del user_row['drop2']

display(rating_row.head(3))
display(movie_row.head(3))
display(user_row.head(3))
```

### 3. 공통 컬럼을 이용한 데이터 병합

```python
# 3. 공통 컬럼을 이용한 데이터 병합
# merge()
rating_movie_merge = pd.merge(rating_row , movie_row , on='movieId' , how = 'left')
# rating_movie_merge
user_rating_movie_merge = pd.merge(rating_movie_merge , user_row , on='userId' , how = 'left')
user_rating_movie_merge
```

### 4. user 평점 갯수 순위 확인(sort)

```python
# 4. user 평점 갯수 순위 확인(sort)
# user_rating_movie_merge['userId'].value_counts()
temp_dataframe = pd.DataFrame({
    'user_id_cnt' : user_rating_movie_merge['userId'].value_counts()
})
temp_dataframe.index.name = 'userId'
temp_dataframe = temp_dataframe.reset_index()

temp_dataframe['user_id_cnt_rank'] = temp_dataframe['user_id_cnt'].rank(ascending=False).astype(int)

result_merge = pd.merge(user_rating_movie_merge , temp_dataframe , on='userId' , how = 'left')
display( result_merge )
```

### 5. 각 영화의 성별 평균 평점

```python
# 5. 각 영화의 성별 평균 평점
gender_grouped = result_merge['rating'].groupby(result_merge['gender'])
gender_grouped.mean()
```

### 6.  평점이 220개 이상인 영화만 필터링(filter())

```python
# 영화별 평점 개수 확인
temp_dataframe = pd.DataFrame({
    'movieId_cnt' : user_rating_movie_merge['movieId'].value_counts()
})
temp_dataframe.index.name = 'movieId'
temp_dataframe = temp_dataframe.reset_index()

movie_cnt_df = pd.merge(user_rating_movie_merge , temp_dataframe , on='movieId' , how = 'left')
display( movie_cnt_df )
```

```python
# movieId_cnt 220 이상인 데이터 필터링
cnt_220 = movie_cnt_df[movie_cnt_df.movieId_cnt >= 220]
cnt_220
```

**6-2 (추가) 평점이 220개 이상인 영화들의 각 영화별 평균 평점을 구하고, 영화 목록을 중복 없이 표시하기**
- 평점은 내림차순 정렬

```python
movie_grouped = result_merge['rating'].groupby(result_merge['movieId'])
movie_grouped.mean()
```

```python
result2_merge = pd.merge(result_merge , movie_grouped.mean() , on='movieId' , how = 'left')
display( result2_merge )
```

```python
# 영화 목록 (ID, 제목, 장르, 리뷰 수, 평균 평점)
result2_merge = result2_merge.drop_duplicates(['movieId'])
result2_sorted = result2_merge.sort_values(['rating_y'], ascending=False)
list_by_rating = result2_sorted[['movieId', 'title', 'genres', 'user_id_cnt', 'rating_y']]
list_by_rating
```

### 7. 평점이 200개 이상인 영화들의 성별 평균 평점

```python
# 7. 평점이 200개 이상인 영화들의 성별 평균 평점
temp_dataframe = pd.DataFrame({
    'movie_id_cnt' : result_merge['movieId'].value_counts()
})
temp_dataframe.index.name = 'movieId'
temp_dataframe = temp_dataframe.reset_index()
result_movie_merge = pd.merge(result_merge , temp_dataframe , on='movieId' , how = 'left')
result_movie_merge

over_200_movie = result_movie_merge[ result_movie_merge['movie_id_cnt'] >= 200 ]
# display( over_200_movie.head() )
gender_m = over_200_movie[ over_200_movie['gender'] == 'M']['rating'].mean()
gender_f = over_200_movie[ over_200_movie['gender'] == 'F']['rating'].mean()

print('남자 성별 평균 평점 - ',gender_m)
print('여자 성별 평균 평점 - ',gender_f)

```

### 8. 여성이 가장 좋아하는 영화 순위

```python
female_df = result_merge[result_merge.gender == 'F']
f_mean_ratings = female_df['rating'].groupby(result_merge['movieId']).mean()
female_merge = pd.merge(female_df, f_mean_ratings, on='movieId', how='left')
display( female_merge )
```

```python
# 영화 목록 (ID, 제목, 장르, 리뷰 수, 평균 평점)
female_merge = female_merge.drop_duplicates(['movieId']) # 영화 값 중복 제거
female_sorted = female_merge.sort_values(['rating_y'], ascending=False) # 평점 순 정렬
female_list_by_rating = female_sorted[['movieId', 'title', 'genres', 'user_id_cnt', 'rating_y']]
female_list_by_rating = female_list_by_rating.reset_index()
female_list_by_rating.index = female_list_by_rating.index+1
del female_list_by_rating['index']
female_list_by_rating
```

### 9. 남녀간의 평균 평점 차이

```python
# 9. 남녀 간 평균 평점 차이
temp_dataframe = pd.DataFrame({
    'movie_id_cnt' : result_merge['movieId'].value_counts()
})
temp_dataframe.index.name = 'movieId'
temp_dataframe = temp_dataframe.reset_index()
result_movie_merge = pd.merge(result_merge , temp_dataframe , on='movieId' , how = 'left')
result_movie_merge

gender_m = result_movie_merge[ result_movie_merge['gender'] == 'M']['rating'].mean()
gender_f = result_movie_merge[ result_movie_merge['gender'] == 'F']['rating'].mean()

print('남자 성별 평균 평점 - ',gender_m)
print('여자 성별 평균 평점 - ',gender_f)
print('남녀 간 평균 평점 차이 - ', gender_f - gender_m)
```
