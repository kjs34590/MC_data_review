```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds')
print('response : ', response)
```

```python
# 통째로 하나의 문자열로 출력됨
response.text
```

- Parsing할 수 있는 형태로 만들기

```python
import re
# response.text
soup = BeautifulSoup(response.text , 'lxml')
rows = soup.select('div > ul > li')
print('type - ' , type(rows))
print()

etfs = {}
etf_name   = []
etf_market = []
etf_ticker = []
for row in rows : 
    # print(row.text)
    # print("*"* 100)
    try : 
        etf_name = re.findall('^(.*) \(NYSE' , row.text)
        # print(etf_name)
        etf_market = re.findall('\((.*)\|' , row.text)
        # print(etf_market)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)' , row.text)
        # print(etf_ticker)
        if ( (len(etf_name) > 0 ) & (len(etf_market) > 0) & (len(etf_ticker) > 0) ) :
            etfs[ etf_ticker[0] ] = [ etf_market[0] , etf_name[0]]
    except AttributeError as err :
        pass
```

```python
temp_frm = pd.DataFrame(etfs)
temp_frm
```

#### 데이터프레임을 이용해 행, 열 단위의 데이터 가공하기

- 람다식, 인자로 함수를 받을 수 있다

```python
def get_square(a) : 
    return a ** 2
```

```python
print(' - ', get_square(2) )
```

```python
print('lambda - ')
lambda_square = lambda a: a ** 2
print(' - ', lambda_square(2))
```

#### 타이타닉 실습 [2]

```python
import seaborn as sns
datasets = sns.load_dataset('titanic')
print('type - ')
print(type(datasets))
```

```python
datasets.head()
```

```python
# age를 기준으로 성인/미성년자 구분
datasets['adult'] = datasets.apply(lambda frm : 'adult' if frm['age']>=20 else 'child', axis=1)
datasets.head()
```

```python
datasets['category'] = datasets.apply(lambda frm : frm['sex'] if frm ['age'] >= 20 else 'child', axis=1)
datasets.tail()
```

```python
# apply는 Series 타입에도 적용 가능 -> 축 방향 설정 필요 없음
datasets['embarked_len'] = datasets['embark_town'].apply(lambda frm : len(str(frm)))
datasets.head()
```

- 15세 이하는 child ,  15~60 까지는 adult , 61 이상은 elderly 로 분류를해서 age_division 컬럼에 저장
- lambda 식으로 apply함수를 이용해서 구현한다면

```python
datasets['age_division'] = datasets['age'].apply(lambda frm : 'child' if frm<=15 else 'adult' if frm <= 60 else 'elderly')
datasets.head()
```

```python
datasets['age_division'].value_counts()
```

```python
datasets.info()
```

- 승객에 대한 나이(age)와 성별(sex)에 따른 카테고리를 'age/gender' 추가한다
- 조건 1 : 20살이 넘으면 성별을 그대로 사용하고
- 조건 2 : 20살 미만이면 성별에 관계없이 'child' 라고 정의하고 결측값이 있는 경우는 'i don't know'정의한다

```python
# 결측값 고려(isnull 미사용)
datasets['age/gender'] = datasets.apply(lambda frm : frm['sex'] if frm['age']>=20 else 'child' if frm['age'] < 20 else "I don't know", axis=1)
datasets.tail()
```

```python
# 결측값 고려 (isnull 사용)
datasets['age/gender'] = datasets.apply(lambda frm : frm['sex'] if frm['age']>=20 else "I don't know" if pd.isnull(frm['age']) else 'child', axis=1)
datasets.tail(30)
```

```python
# 결측값 고려 (isnull사용, lambda식 정의)
condition01 = lambda frm : frm['sex'] if frm['age'] >= 20 else 'child'
condition02 = lambda frm : "I don't know" if pd.isnull(frm['age']) else condition01(frm)

datasets['age/gender'] = datasets.apply(condition02, axis=1)
datasets.tail(20)
```

```python
print('나이가 알려지지 않은 승객의 나이를 평균 나이로 대체하기')
datasets['age'] = datasets['age'].fillna(datasets['age'].mean())
datasets.tail(20)
```

- age_gender_category 컬럼 추가하고자 한다
- 조건 1 : 성별(male | female)
- 조건 2 : 성별을 나타내는 문자열 뒤에 나이를 추가하고 싶다

#### 실습
- 10명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.
- 1. 학생 이름을 나타내는 열을 포함시키지 않고 
- 데이터프레임 df_score1 을 생성한 후, 
- df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
- reset_index 명령으로 이 인덱스 열을 명령으로 
- 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

- 2.
- 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
- set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.


```python
data = {
    '국어' : [50, 65, 87, 10, 100, 89, 32, 83, 100, 58],
    '영어' : [100, 30, 20, 11, 88, 97, 100, 100, 10, 57],
    '수학' : [100, 59, 88, 29, 99, 29, 22, 99, 29, 50]
}
```

```python
score_frm = pd.DataFrame(data)
score_frm
```

```python
score_frm.index = ['려명', '승윤', '태형', '강희', '지수', '미영', '수연', '홍진', '예은', '신옥']
score_frm.index.name = '이름'
score_frm
```

```python
r_score_frm = score_frm.reset_index()
r_score_frm
```

```python
s_score_frm = r_score_frm.set_index('이름')
s_score_frm
```
